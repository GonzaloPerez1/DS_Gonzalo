### Realiza un análisis de encuestados por género con visualización.

### ¿Cuales son los 15 paises más encuestados? Hazlo con visualización

```python
df_7 = pd.DataFrame(multi_choice['Country'].value_counts().sort_values(ascending = False))
df_7.reset_index(inplace=True)
df_7 = df_7[df_7['index'] != 'Other']
df_7 = df_7.iloc[:14,:]

plt.figure(figsize=(10,7))
plt.hlines(y=df_7['index'],
           xmin=0,
           xmax=df_7['Country'],
           color='#A300F5')
plt.plot(df_7['Country'], df_7['index'], "o", color='#A300F5');
```

### Realiza un análisis de salarios por género con visualización.

```python
multi_choice.head()
```

```python
divisas = pd.read_csv('conversionRates.csv')
divisas = divisas.drop('Unnamed: 0', axis = 1)
divisas.head()
```

```python
multi_choice.columns.tolist()
```

```python
#multi_choice['CompensationCurrency'].notna()
df_8 = multi_choice[multi_choice['CompensationCurrency'].notna()]
df_8['CompensationCurrency']

df_total = pd.merge(df_8, divisas, left_on='CompensationCurrency', right_on='originCountry')
df_total = df_total[['GenderSelect', 'CompensationAmount', 'CompensationCurrency', 'exchangeRate']]
df_total = df_total[df_total['CompensationAmount'].notna()]

df_total['CompensationAmount'] = df_total['CompensationAmount'].apply(lambda x: x.replace(',', ''))
df_total['CompensationAmount'] = df_total['CompensationAmount'].apply(lambda x: x.replace('-', '0'))
df_total['CompensationAmount'] = df_total['CompensationAmount'].astype(float)

df_total['real_money'] = df_total['CompensationAmount'] * df_total['exchangeRate']
df_total
```

```python
for i, j in enumerate(df_total['CompensationAmount']):
    #print(i)
    if not isinstance(j, str):
        print(j)
        print(type(j))
```

```python
isinstance(j, str)
```

```python
graph =pd.DataFrame(df_total.groupby('GenderSelect')['real_money'].median())

graph.plot(kind='bar', stacked=True)
```
