# insertamos datos en la BBDD

```python
import sqlite3
import pandas as pd
```

```python
#Leemos los datasets
hospitales = pd.read_csv('./data/catalogo_hospitales.csv')
equipo_medico = pd.read_csv('./data/Equipo_Médico.csv', sep='\t')
gasto_sanidad = pd.read_csv('./data/gasto_sanidad_ccaa_2019.csv', sep=';')
pop_mujeres = pd.read_csv('./data/mujeres_2019.csv', sep=';')
pib = pd.read_csv('./data/pib_ccaa_2019.csv', sep=';')
poblacion = pd.read_csv('./data/pop_sup_2019.csv', sep=';')
```

```python
#Abrimos al conexión con la base de datos
conn = sqlite3.connect('database_eda')
cursor = conn.cursor()
```

```python
#Insertamos los datos
hospitales.to_sql('HOSPITALES', conn, if_exists='append', index=False)
equipo_medico.to_sql('EQUIPO_MEDICO', conn, if_exists='append', index=False)
gasto_sanidad.to_sql('GASTO_SANIDAD', conn, if_exists='append', index=False)
pop_mujeres.to_sql('POP_MUJERES', conn, if_exists='append', index=False)
pib.to_sql('PIB', conn, if_exists='append', index=False)
poblacion.to_sql('POBLACION', conn, if_exists='append', index=False)
```

## Realizamos las queries

```python
#10 hospitales con mayor número de camas
query = '''
SELECT NOMBRE, NUMERO_DE_CAMAS
FROM HOSPITALES
ORDER BY NUMERO_DE_CAMAS DESC
LIMIT 10;
'''

df1 = pd.read_sql_query(query, conn)

import numpy as np

def show_values_on_bars(axs, h_v="v", space=0.3):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)


#Grafica
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
g = sns.barplot(
    x = 'NUMERO_DE_CAMAS',
    y = 'NOMBRE',
    palette = 'magma',
    data = df1
);

show_values_on_bars(g, 'h', 0.3)

plt.tight_layout() #sirve apra que no corte bordes
plt.savefig('hospitales_mas_camas.jpeg', dpi=100)
```

```python
#NUMERO DE MAMOGRAFOS POR MUJERES Y CCAA
query = '''
SELECT E.COMUNIDAD_AUTONOMA, MAMOS, NUM_MUJERES
FROM EQUIPO_MEDICO E
JOIN POP_MUJERES M ON E.COMUNIDAD_AUTONOMA = M.COMUNIDAD_AUTONOMA
ORDER BY MAMOS DESC;
'''

df2 = pd.read_sql_query(query, conn)
df2
```

```python
df2['RATIO_MAMOS'] = df2['NUM_MUJERES']/df2['MAMOS']
```

```python
import numpy as np

def show_values_on_bars(axs, h_v="v", space=0.3):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)

#Grafica
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
sns.barplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'MAMOS',
    data = df2
)
sns.barplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'RATIO_MAMOS',
    data = df2
);


plt.xticks(rotation=-45);
#show_values_on_bars(g, 'h', 0.5)

#plt.tight_layout() #sirve apra que no corte bordes
#plt.savefig('hospitales_mas_camas.jpeg', dpi=100)
```

```python
#NUMERO DE CAMAS POR POBLACION Y CCAA (CALIDAD ASISTENCIAL)
query = '''
SELECT H.COMUNIDAD_AUTONOMA, SUM(H.NUMERO_DE_CAMAS) AS NUMERO_DE_CAMAS, P.POBLACION
FROM HOSPITALES H
JOIN POBLACION P ON H.COMUNIDAD_AUTONOMA = P.COMUNIDAD_AUTONOMA
GROUP BY H.COMUNIDAD_AUTONOMA
ORDER BY P.POBLACION DESC;
'''

df3 = pd.read_sql_query(query, conn)
df3
```

```python
query = '''
SELECT SUM(NUMERO_DE_CAMAS) AS CAMAS
FROM HOSPITALES;
'''

camas_totales = pd.read_sql_query(query, conn)
camas_totales
```

```python
query = '''
SELECT SUM(POBLACION)
FROM POBLACION;
'''

poblacion_total = pd.read_sql_query(query, conn)
poblacion_total
```

```python
df3['PER_POB'] = (df3['POBLACION']*100)/poblacion_total['SUM(POBLACION)']
df3['PER_CAMA'] = (df3['NUMERO_DE_CAMAS']*100)/camas_totales['CAMAS']
```

```python
for i in range(len(df3)):
    df3['PER_CAMA'][i] = (df3['NUMERO_DE_CAMAS'][i]*100)/camas_totales['CAMAS'] #PORCENTAJE DE CAMAS POR CCAA
    df3['PER_POB'][i] = (df3['POBLACION'][i]*100)/poblacion_total['SUM(POBLACION)'] #PORCENTAJE DE POBLACIÓN POR CCAA
df3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,12))
sns.barplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'POBLACION',
    color = 'skyblue',
    data = df3
)

#AÑADIR PIB POR COMUNIDAD AUTÓNOMA

plt.xticks(rotation=80);

plt.twinx()

sns.lineplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_CAMA',
    color = 'blueviolet',
    label = 'Porcentaje de camas por CCAA',
    data = df3
)

sns.scatterplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_CAMA',
    color = 'blueviolet',
    data = df3
)

sns.lineplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_POB',
    color = 'darkblue',
    label = 'Porcentaje de población por CCAA',
    data = df3
)

sns.scatterplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_POB',
    color = 'darkblue',
    data = df3
)

plt.ylim(0,100);
plt.legend(loc = 'best', fontsize='x-large');

plt.tight_layout() #sirve apra que no corte bordes
plt.savefig('poblacion_camas.jpeg', dpi=100)
```

```python
query = '''
SELECT H.COMUNIDAD_AUTONOMA, SUM(H.NUMERO_DE_CAMAS) AS NUMERO_DE_CAMAS, P.POBLACION, PI.PIB 
FROM HOSPITALES H
JOIN POBLACION P ON H.COMUNIDAD_AUTONOMA = P.COMUNIDAD_AUTONOMA
JOIN PIB PI ON P.COMUNIDAD_AUTONOMA = PI.COMUNIDAD_AUTONOMA
GROUP BY H.COMUNIDAD_AUTONOMA
ORDER BY PI.PIB DESC;
'''

df3 = pd.read_sql_query(query, conn)
df3

df3['PER_POB'] = (df3['POBLACION']*100)/poblacion_total['SUM(POBLACION)']
df3['PER_CAMA'] = (df3['NUMERO_DE_CAMAS']*100)/camas_totales['CAMAS']

for i in range(len(df3)):
    df3['PER_CAMA'][i] = (df3['NUMERO_DE_CAMAS'][i]*100)/camas_totales['CAMAS'] #PORCENTAJE DE CAMAS POR CCAA
    df3['PER_POB'][i] = (df3['POBLACION'][i]*100)/poblacion_total['SUM(POBLACION)'] #PORCENTAJE DE POBLACIÓN POR CCAA
df3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,12))
sns.barplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PIB',
    color = 'skyblue',
    data = df3
)

#AÑADIR PIB POR COMUNIDAD AUTÓNOMA

plt.xticks(rotation=80);

plt.twinx()

sns.lineplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_CAMA',
    color = 'blueviolet',
    label = 'Porcentaje de camas por CCAA',
    data = df3
)

sns.scatterplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_CAMA',
    color = 'blueviolet',
    data = df3
)

sns.lineplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_POB',
    color = 'darkblue',
    label = 'Porcentaje de población por CCAA',
    data = df3
)

sns.scatterplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PER_POB',
    color = 'darkblue',
    data = df3
)

plt.ylim(0,100);
plt.legend(loc = 'best', fontsize='x-large');

plt.tight_layout() #sirve apra que no corte bordes
plt.savefig('pib_camas.jpeg', dpi=100)
```

```python
#GASTO EN SANIDAD SOBRE EL PIB
query = '''
SELECT G.COMUNIDAD_AUTONOMA, G.GASTO_SANIDAD, P.PIB
FROM GASTO_SANIDAD G
JOIN PIB P ON G.COMUNIDAD_AUTONOMA = P.COMUNIDAD_AUTONOMA
GROUP BY G.COMUNIDAD_AUTONOMA
ORDER BY P.PIB DESC;
'''

df4 = pd.read_sql_query(query, conn)
df4
```

```python
df4['GASTO_SANIDAD_PIB'] = df4['GASTO_SANIDAD']/df4['PIB']
df4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,12))
sns.barplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'PIB',
    data = df4
)

plt.xticks(rotation=80);

plt.twinx()

sns.lineplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'GASTO_SANIDAD_PIB',
    data = df4
);

sns.scatterplot(
    x = 'COMUNIDAD_AUTONOMA',
    y = 'GASTO_SANIDAD_PIB',
    data = df4
);
```
