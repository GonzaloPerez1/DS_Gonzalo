## 14. Cuantos clientes hay de españa y de Brazil

```python
query= '''
SELECT COUNTRY, COUNT(*)
FROM CUSTOMERS
WHERE COUNTRY = 'Brazil' OR COUNTRY = 'Spain'
GROUP BY COUNTRY;
'''

cursor.execute(query)
cursor.fetchall()
```

## 9. Quien ha sido el agente de ventas que más ha vendido en 2009?

```python
query='''
SELECT E.FIRSTNAME, SUM(I.TOTAL)
FROM CUSTOMERS C
JOIN EMPLOYEES E ON C.SUPPORTREPID = E.EMPLOYEEID
JOIN INVOICES I ON C.CUSTOMERID = I.CUSTOMERID
WHERE I.INVOICEDATE LIKE '2009%'
GROUP BY E.FIRSTNAME
ORDER BY SUM(I.TOTAL) DESC;
'''

cursor.execute(query)
cursor.fetchall()
```

```python
query='''
SELECT E.FIRSTNAME, SUM(I.TOTAL)
FROM CUSTOMERS C
JOIN EMPLOYEES E ON C.SUPPORTREPID = E.EMPLOYEEID
JOIN INVOICES I ON C.CUSTOMERID = I.CUSTOMERID
WHERE I.INVOICEDATE LIKE '2009%'
GROUP BY E.FIRSTNAME
ORDER BY SUM(I.TOTAL) DESC;
'''
df = pd.read_sql_query(query, conn)
df
```

## 10. Quienes son los 3 grupos que más han vendido?

```python
query='''
SELECT AR.NAME, COUNT(T.TRACKID)
FROM ARTISTS AR
JOIN ALBUMS AL ON AR.ARTISTID = AL.ARTISTID
JOIN TRACKS T ON AL.ALBUMID = T.ALBUMID
JOIN INVOICE_ITEMS IT ON T.TRACKID = IT.TRACKID
JOIN INVOICES I ON IT.INVOICEID = I.INVOICEID
GROUP BY AR.NAME
ORDER BY COUNT(I.TOTAL) DESC
LIMIT 3;
'''

cursor.execute(query)
cursor.fetchall()
```

```python
query='''
SELECT AR.NAME, COUNT(T.TRACKID)
FROM ARTISTS AR
JOIN ALBUMS AL ON AR.ARTISTID = AL.ARTISTID
JOIN TRACKS T ON AL.ALBUMID = T.ALBUMID
JOIN INVOICE_ITEMS IT ON T.TRACKID = IT.TRACKID
JOIN INVOICES I ON IT.INVOICEID = I.INVOICEID
GROUP BY AR.NAME
ORDER BY COUNT(I.TOTAL) DESC
LIMIT 3;
'''
df = pd.read_sql_query(query, conn)
df
```
