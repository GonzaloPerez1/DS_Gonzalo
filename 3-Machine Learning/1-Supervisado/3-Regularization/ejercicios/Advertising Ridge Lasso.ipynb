## Import libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Exploratory Data Analysis 

```python
DATAPATH = 'data/Advertising.csv'

data = pd.read_csv(DATAPATH)
data.drop(columns=['Unnamed: 0'], inplace=True)
data.head()
```

```python
data.shape
```

```python
#Vemos la distribución de los datos
import seaborn as sns

sns.pairplot(data);
```

```python
#Comprobamos las relaciones existentes entre features
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(),
           vmin = -1,
           vmax = 1,
           cmap = sns.diverging_palette(145, 280, s=85, l=25, n=7),
           annot = True);
```

## Comenzamos con el modelo
1. Dividimos los datos.
2. Creamos las partes de Train y Test.
3. Estandarizamos los datos (Este paso es opcional).

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Separamos features y target
X = data[['TV', 'radio', 'newspaper']]
y = data['sales']

#Separamos train y test
X_train, x_test, Y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 42)

#Estandarizamos los datos
scaler = StandardScaler()
scaler.fit(X_train) #"Entrenamos" el escalado

X_train_scal = scaler.transform(X_train) #Features ya escaladas
x_test_scal = scaler.transform(x_test) #Target ya escalado
```

## Regresión Lineal

```python
from sklearn import metrics
from sklearn.linear_model import LinearRegression

#Creamos un objeto LinearRegression
lm = LinearRegression()
lm.fit(X_train_scal, Y_train) #Entrenamos el modelo

#Predecimos los valores
y_pred = lm.predict(x_test_scal)

#Comprobamos el error que hay entre el target real y el predicho
baseline_error = metrics.mean_squared_error(y_test, y_pred)

#Comprobamos los errores cometidos en train y test
print('Train MSE: ', metrics.mean_squared_error(Y_train, lm.predict(X_train_scal)))
print('Test MSE: ', metrics.mean_squared_error(y_test, lm.predict(x_test_scal)))
```

## Regularización
### Ridge

```python
from sklearn.linear_model import Ridge

#Creamos el objeto Ridge con alpha aleatoria (1)
ridgeR = Ridge(alpha = 1)
ridgeR.fit(X_train_scal, Y_train) #Entrenamos el modelo con los datos escalados

#Comprobamos los errores cometidos en el modelo
print('Train MSE: ', metrics.mean_squared_error(Y_train, ridgeR.predict(X_train_scal)))
print('Test MSE: ', metrics.mean_squared_error(y_test, ridgeR.predict(x_test_scal)))
```

### Buscamos el mejor valor para alpha

```python
n_alphas = 1000 #Número de cortes del eje X
alphas = np.logspace(-4, 3, n_alphas) #Valor del alpha

coef_ridge = []
err_ridge = []
baseline = []
alpha = []

for a in alphas:
    alpha.append(a) #Usaremos la lista para buscar el valor mínimo
    
    ridge = Ridge(alpha = a)
    ridge.fit(X_train_scal, Y_train)
    
    coef_ridge.append(ridge.coef_) #Coeficientes (Betas)
    
    y_pred = ridge.predict(x_test_scal)
    ridge_error = metrics.mean_squared_error(y_pred, y_test) #MSE
    
    err_ridge.append(ridge_error)
    baseline.append(baseline_error)
```

```python
from sklearn.model_selectio import GridSearchCV

n_alphas = 500

ridge = Ridge()

parameters = {'alpha': np.logspace(-4, 1, n_alphas)}

ridge_gridsearch = GridSearchCV(ridge, parameters, cv = 5, scoring = 'neg_mean_squared_error') #cv = validacion cruzada de 5

ridge_gridsearch.fit(X_train_scal, Y_train)
```

```python
ridge_gridsearch.best_params_ #Visualizamos el mejor valor de alpha
```

```python
ridge_best = ridge_gridsearch.best_estimator_
```

```python
#Buscaremos el valor de alpha para el menor valor de error
import pandas as pd

df = pd.DataFrame(alpha, columns = ['Alpha'])
df['Error'] = err_ridge
df[df['Error'] == df['Error'].min()]
```

### Visualizamos la recta con los valores de alpha

```python
plt.figure(figsize=(20,12))
ax = plt.gca()
ax.plot(alphas, err_ridge, linewidth=5, color='red', label="Ridge regression")
ax.plot(alphas, baseline, linewidth=4,linestyle='--', color='blue', label='Linear regression')
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('error', fontsize=30)
ax.legend(fontsize=30)
plt.title(r'Regression error ($\lambda$)', fontsize=30)
plt.show()
```

### Probamos de nuevo el modelo con la modificación del alpha

```python
from sklearn.linear_model import Ridge

#Creamos el objeto Ridge con alpha aleatoria (1)
ridgeR = Ridge(alpha = 0.0001) #alpha = ridge_best
ridgeR.fit(X_train_scal, Y_train) #Entrenamos el modelo con los datos escalados

#Comprobamos los errores cometidos en el modelo
print('Train MSE: %0.4f' % metrics.mean_squared_error(Y_train, ridgeR.predict(X_train_scal)))
print('Test MSE: %0.4f' % metrics.mean_squared_error(y_test, ridgeR.predict(x_test_scal)))
```

### Visualizamos la modificación de los coeficientes

```python
plt.figure(figsize=(20,12))
ax = plt.gca()
ax.plot(alphas, coef_ridge, linewidth=5)
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('weight value', fontsize=30)
plt.title('Ridge coefficients paths', fontsize=30)
plt.show()
```

### Lasso

```python
from sklearn.linear_model import Lasso

#Creamos un objeto Lasso
lassoR = Lasso(alpha = 1)
lassoR.fit(X_train_scal, Y_train) #Entrenamos el modelo

#Buscamos el error en train y test
print("Train MSE: %0.4f" % metrics.mean_squared_error(Y_train, lassoR.predict(X_train_scal)))
print("Test MSE: %0.4f" % metrics.mean_squared_error(y_test, lassoR.predict(x_test_scal)))
```

### Buscamos el major valor para alpha

```python
n_alphas = 1000
alphas = np.logspace(-4, 3, n_alphas)

lasso = Lasso(fit_intercept=False)

coef_lasso = []
err_lasso = []
alpha = []

for a in alphas:
    alpha.append(a)
    
    lasso.set_params(alpha=a)
    lasso.fit(X_train_scal, Y_train)
    
    coef_lasso.append(lasso.coef_)
    
    y_pred = lasso.predict(x_test_scal)
    
    lasso_error = metrics.mean_squared_error(y_pred, y_test)    
    err_lasso.append(lasso_error)
```

```python
#Buscaremos el valor de alpha para el menor valor de error
import pandas as pd

df = pd.DataFrame(alpha, columns = ['Alpha'])
df['Error'] = err_lasso
df[df['Error'] == df['Error'].min()]
```

### Visualizamos la recta con los valores de alpha

```python
plt.figure(figsize=(20,10))
ax = plt.gca()
ax.plot(alphas, err_lasso, linewidth=5, color='red', label="Lasso")
ax.plot(alphas, baseline, linewidth=4,linestyle='--', color='blue', label='Linear regression')
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('error', fontsize=30)
ax.legend(fontsize=30)
plt.title(r'Regression error ($\lambda$)', fontsize=30)
plt.show()
```

### Probamos el modelo con la modificación de alpha

```python
from sklearn.linear_model import Lasso

#Creamos un objeto Lasso
lassoR = Lasso(alpha = 0.0001)
lassoR.fit(X_train_scal, Y_train) #Entrenamos el modelo

#Buscamos el error en train y test
print("Train MSE: %0.4f" % metrics.mean_squared_error(Y_train, lassoR.predict(X_train_scal)))
print("Test MSE: %0.4f" % metrics.mean_squared_error(y_test, lassoR.predict(x_test_scal)))
```

### Visualizamos la modificación de las betas

```python
plt.figure(figsize=(20,12))
ax = plt.gca()
ax.plot(alphas, coef_lasso, linewidth=5)
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('weight value', fontsize=30)
plt.title('Lasso coefficients paths', fontsize=30)
plt.show()
```

### Elastic Net

```python
from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet(alpha = 1, l1_ratio = 0.5)
elastic_net.fit(X_train_scal, Y_train)

print("Train MSE: %0.4f" % metrics.mean_squared_error(Y_train, elastic_net.predict(X_train_scal)))
print("Test MSE: %0.4f" % metrics.mean_squared_error(y_test, elastic_net.predict(x_test_scal)))
```

### Ajustamos los valores de alpha y l1_ratio

```python
n_alphas = 1000
alphas = np.logspace(-4, 3, n_alphas)

coef_eln = []
err_eln = []
baseline = []
alpha = []

for a in alphas:
    alpha.append(a)
    
    elastic_net = ElasticNet(alpha = a, l1_ratio=0.8)
    elastic_net.fit(X_train_scal, Y_train)
    
    coef_eln.append(elastic_net.coef_)
    
    y_pred = elastic_net.predict(x_test_scal)
    elasticnet_error = metrics.mean_squared_error(y_pred, y_test)
    
    err_eln.append(elasticnet_error)
    baseline.append(baseline_error)
```

```python
from sklearn.model_selectio import GridSearchCV

n_alphas = 500

elasticNet = ElasticNet()

parameters = {'alpha': np.logspace(-4, 1, n_alphas),
              'l1_ratio': np.arange(0.01, 1, 0.01)}

elastic_gridsearch = GridSearchCV(elasticNet, parameters, cv = 5, scoring = 'neg_mean_squared_error') #cv = validacion cruzada de 5

elastic_gridsearch.fit(X_train_scal, Y_train)
```

```python
elastic_gridsearch.best_params_
```

```python
elastic_best = elastic_gridsearch.best_estimator_ #Esto es el modelo final ya entrenado
```

```python
#Buscaremos el valor de alpha para el menor valor de error
import pandas as pd

df = pd.DataFrame(alpha, columns = ['Alpha'])
df['Error'] = err_eln
df[df['Error'] == df['Error'].min()]
```

### Visualizamos las alphas

```python
plt.figure(figsize=(20,10))
ax = plt.gca()
ax.plot(alphas, err_eln, linewidth=5, color='red', label="Elastic net")
ax.plot(alphas, baseline, linewidth=4,linestyle='--', color='blue', label='Linear regression')
ax.set_xscale('log')
plt.xlabel('$\lambda$', fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel('error', fontsize=30)
ax.legend(fontsize=30)
plt.title(r'Regression error ($\alpha=0.5$)', fontsize=30)
plt.show()
```
